```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Data Preprocessing

```
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None) # becoz there is no column name so to include first row as well we use header = none
#apriori takes the input data ind ifferent format unlike other algorithms so we are creating a list of transactions and appending the data into that list
# we will two for loops one for all rows and one for all column to append each of the string inside the list
transactions = []
for i in range(0, 7501):
  transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])
```

## Training the Apriori model on the dataset

```
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

## Visualising the results

### Displaying the first results coming directly from the output of the apriori function

```
result = list(rules)
```

```
print(result)
```

### Putting the results well organised into a Pandas DataFrame

```
def inspect(results): #Defines a function inspect that processes the Apriori results.
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    #result[2] contains association rule details.
    #result[2][0] is the first rule extracted from the result.
    #result[2][0][0] is the LHS (antecedent) of the rule.
    #tuple(result[2][0][0]) converts it into a tuple.
    #[0] extracts the first item (since it's stored as a set).
    rhs         = [tuple(result[2][0][1])[0] for result in results] # same as rhs
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts)) #zip() combines lhs, rhs, supports, confidences, and lifts into a list of tuples.
resultsinDataFrame = pd.DataFrame(inspect(result), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```
